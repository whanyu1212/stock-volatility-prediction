```python
import pandas as pd
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer
```

```python
# Define the model path
model_name = "ProsusAI/finbert"

# Load tokenizer and model from Hugging Face Hub
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create a pipeline for sentiment analysis
finbert_sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
news = pd.read_csv("news_data.csv")
news
```

```python
# Drop rows with empty cells and reset index
news.dropna(inplace=True)
news.reset_index(drop=True, inplace=True)
```

```python
news
```

```python
# Initialize an empty list to store the sentiment scores
sentiment_scores = []

# Analyze the sentiment of all financial texts and store the scores
for index, row in news.iterrows():
    text = row['text']
    sentiment = finbert_sentiment_pipeline(text)
    # Extract the sentiment score (assuming the first element of the result is the sentiment label)
    sentiment_score = sentiment[0]['score']
    print(index,sentiment_score)
    sentiment_scores.append(sentiment_score)
```

```python
# Add the sentiment scores to the DataFrame as a new column
news['sentiment_score'] = sentiment_scores
```

```python
news
```

```python
news.to_csv("processed_news_data.csv", index=False)
```

```python
# Convert the 'publishedDate' column to datetime
news['publishedDate'] = pd.to_datetime(news['publishedDate'])
# Extract the date from the 'publishedDate' column
news['date'] = news['publishedDate'].dt.date
news
```

```python
# Calculate the median sentiment score for each date
median_scores = news.groupby('date')['sentiment_score'].median().reset_index()
median_scores
```

```python
# Save the DataFrame with median scores to a CSV file
median_scores.to_csv("median_sentiment_scores.csv", index=False)
```
